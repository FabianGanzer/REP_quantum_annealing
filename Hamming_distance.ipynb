```python
import numpy as np
from haming_distance_distribution import hamming_distance_distribution
import matplotlib.pyplot as plt
```

```python
N_repeat = 200

N = 5               # number of users
M = 4               # length of id-sequence for every user
K = 100              # number of antennas
xi = 0              # std of thermal noise

which_ctl_fct = 0   # 0: linear control function, 1: optimal control function
neglection_rule = 1 # 0: neglect only smallest matrix element, 1: neglect everything below thresold
neglection_thres_list = [0, 0.01, 0.1, 0.2, 0.4, 0.6, 0.8, 1]  # thres only used if reglection_rule == 1

nb_pts_gap = 20     # number of points for the gap computation
nb_pts_time = 30    # number of points for resolution of the time dependant Schrodinger's equation
epsilon = 0.1       # precision level for the control function (valid for both, linear and optimal scheduling)
gamma = 1           # strength of the transverse field, irrelevant for us 

resol = 100
height_plt = 6
width_plt = height_plt

# ----------- Program -----------
# activity pattern
alpha = np.zeros(N)
alpha[0] = 1
alpha[2] = 1
```

```python
print(f"per diagram {N_repeat} samples")
for k in range(len(neglection_thres_list)):
    neglection_thres = neglection_thres_list[k]

    print(f"neglection thresold: {neglection_thres}")
    d = np.arange(N+1)
    n = np.zeros(N+1)

    d, n = hamming_distance_distribution(N_repeat, N, M, alpha, K, xi, 1, neglection_thres, gamma, epsilon, which_ctl_fct, nb_pts_gap, nb_pts_time, False)
    
    plt.bar(d, n/N_repeat)
    plt.xlabel(r"$d_{Hamming}$")
    plt.ylabel(r"$n/N_{r}$")
    plt.ylim([0, 1.1])
    plt.title(rf"$\eta = {neglection_thres}$")
    plt.show()
```
